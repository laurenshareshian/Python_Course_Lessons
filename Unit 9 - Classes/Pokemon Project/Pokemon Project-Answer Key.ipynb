Pokémon types dictate whether one Pokémon's attack is strong or weak (i.e., super effective or not very effective) against another, dealing additional or reduced damage as a result. What is the modifier in the damage formula above? Our simplified modifier will be:

$$ Modifier = Critical \times Random \times Type $$

**Random** is a random integer percentage between 0.85 and 1.00 (inclusive).

Recall that **Type** can be 0.5 (not very effective), 1 (normally effective), or 2 (super effective), depending on how effective the move's type is against the defending Pokémon's type.

Actually, the type can also take on the values ¼ and 4. For targets that have two types (every Pokémon is either one or two types ), such as Bulbasaur, who is both a Grass and Poison type, overall type effectiveness is the product of the effectiveness against each of its types:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is super effective against both of the opponent's types (such as Dig, a Ground-type move, used against an Aggron, a Steel/Rock Pokémon), then the move does 2x2=4 times the damage;

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is not very effective against both of the opponent's types (such as Wake-Up Slap, a Fighting-type move, used against a Sigilyph, a Psychic/Flying Pokémon), then the move only does 1/2x1/2=¼ of the damage;

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-If the type of a move is super effective against one of the opponent's types but not very effective against the other (such as Razor Leaf, a Grass-type move, used against a Gyarados, a Water/Flying Pokémon), then the move deals normal damage 2x1/2=1;

**Critical** is 2 if you have landed a critical hit and 1 otherwise. What is a critical hit? This is of the one last things to understand. A critical hit is a random event that multiplies the damage of a damage-dealing move. When one occurs, the game will display the message "A critical hit!" after the damage has been dealt. 

How does one calculate the probability that the attack was critical? The probability that an attacking Pokémon landed a critical hit based on that attacking Pokémon's speed is: $$ P = Speed / 512 $$

Thus, one can generate a random  number between 0 and 511 and if that number is less than the attacking Pokémon's speed then Critical=2.

Now let's return to levels. When a Pokémon beats another Pokémon, he gains their experience points. With enough experience points, the Pokémon achieves the next level. Levels go from 1 to 100. When a player just starts out, they have 0 experience points and are on level 1. Once they gain experience points, the function that models experience points versus level is the cube root function:

$$ Level = (Experience)^{1/3} $$

We can now take the level of the attacker into account when calculating damage:

$$Damage=(\frac{(\frac{2 \times Level}{5}+2) \times Power \times \frac{A}{D}}{50}) \times Modifier$$

```python
import random as rd

class Pokemon:
    def __init__(self, name):
        self.name = name
        self.type_ = CHARACTERS[name]['Type']
        self.HP = CHARACTERS[name]['HP']
        self.moves = CHARACTERS[name]['Moves']
        self.attack = CHARACTERS[name]['Attack']
        self.defense = CHARACTERS[name]['Defense']
        self.speed = CHARACTERS[name]['Speed']
        self.experience = CHARACTERS[name]['Experience']
        self.level = 1

    def Type_coefficient(self, defender, move):
        '''determines type coefficient in damage calculation'''
        Type = 1
        for type_ in defender.type_:
            if type_ in move['super effective against']:
                Type = 2.0 * Type
            if type_ in move['not very effective against']:
                Type = 0.5 * Type
        return Type

    def critical_coefficient(self, attacker):
        '''determines whether the hit was critical or not based on the attackers speed'''
        if rd.randint(0,512) < attacker.speed:
            critical = 2
            print(f'{attacker.name} landed a critical hit!')
        else:
            critical = 1
        return critical

    def calculate_damage(self, attacker, defender, move):
        '''damage calculator based on attacking and defending pokemon stats'''
        random = rd.randint(85,100)/100
        critical = self.critical_coefficient(attacker)
        Type = self.Type_coefficient(defender, move)
        damage = (2*attacker.level/5 + 2)*move['power']*(attacker.attack/defender.defense)/50*critical*random*Type
        return damage

    def update_level(self, enemy):
        '''update level based on new experience'''
        self.experience += enemy.experience
        self.level = int((self.experience)**(1.0/3.0))
        print(f'Your new level: {self.level} \n')
        print(f'******************************** \n')
        
    def choose_optimal_move(self, defender):
        '''computer choose optimal move to use based on user's pokemons strengths and weaknesses and power of move'''

    #    computer_choice = 'Tackle' # for testing
        move_choices = []
        for move in self.moves:
            move_choices.append(
                (self.Type_coefficient(defender, MOVES_DICTIONARY[move])*MOVES_DICTIONARY[move]['power'], move))
        move_choices.sort(reverse=True)
        computer_choice = move_choices[0][1]
        return MOVES_DICTIONARY[computer_choice]

    def choose_move(self):
        '''Asks user which move to use based on their Pokemon.'''

        move_choice = input(f"Which move do you want to choose out of {pokemon.moves}? ").lower()
#        move_choice = 'bubble' #for testing

    #   if move choice not valid ask user for another
        if move_choice not in [move.lower() for move in self.moves]:
            return choose_move(pokemon)
        else:
            return MOVES_DICTIONARY[move_choice.title()]
        
    def battle(self, enemy, first, optimal_move):
        '''battle until one pokemon dies'''
        
        #if you die:
        if self.HP <= 0:
            print(f'You fainted!')
            
        #if you win:
        elif enemy.HP <= 0:
            print(f'You beat {enemy.name}!')
            self.update_level(enemy)
            
        #battle until someone dies
        else:
            
            #computer always choose optimal move
            enemy_move = enemy.choose_optimal_move(pokemon)
            
            #user attacks computer
            if first == True:
                #user either chooses move or lets computer choose optimal move for them
                if optimal_move:
                    user_move = self.choose_optimal_move(enemy)
                else:
                    user_move = self.choose_move()
                
                #computer takes damage
                damage = self.calculate_damage(self, enemy, user_move)
                enemy.HP = enemy.HP - damage
                print(f"After your {user_move['name']} attack worth {damage :.2f} points: Your HP: {self.HP :.2f} Enemy HP: {enemy.HP:.2f}")
                
                #figure out if computer died or battle again
                self.battle(enemy, False, optimal_move)
                
            #computer attacks user
            else:
                #user takes damage
                damage = self.calculate_damage(enemy, self, enemy_move)
                self.HP = self.HP - damage
                print(f"After your enemy's {enemy_move['name']} attack worth {damage :.2f} points: Your HP: {self.HP:.2f} Enemy HP: {enemy.HP:.2f}")
                
                #figure out if user died or battle again
                self.battle(enemy, True, optimal_move)


    def __str__(self):
        fstring = f"{self.name} is a {self.type_} type with speed {self.speed} and HP {self.HP}.\n"
        fstring += f"{self.name} has moves {self.moves}.\n"
        for move in self.moves:
            fstring += f"{move} with {MOVES_DICTIONARY[move]['power']} power is super effective against the types: "
            fstring += f"{MOVES_DICTIONARY[move]['super effective against']}.\n"
            fstring += f"{move} is not very effective against the types: "
            fstring += f"{MOVES_DICTIONARY[move]['not very effective against']}.\n"
        fstring += f'\n'
        return fstring

class Pikachu(Pokemon):
    def __init__(self):
        super().__init__("Pikachu")

class Charizard(Pokemon):
    def __init__(self):
        super().__init__("Charizard")
        
class Squirtle(Pokemon):
    def __init__(self):
        super().__init__("Squirtle")
        
class Jigglypuff(Pokemon):
    def __init__(self):
        super().__init__("Jigglypuff")
        
class Gengar(Pokemon):
    def __init__(self):
        super().__init__("Gengar")
        
class Magnemite(Pokemon):
    def __init__(self):
        super().__init__("Magnemite")

class Bulbasaur(Pokemon):
    def __init__(self):
        super().__init__("Bulbasaur")
        
class Charmander(Pokemon):
    def __init__(self):
        super().__init__("Charmander")
        
class Beedrill(Pokemon):
    def __init__(self):
        super().__init__("Beedrill")
        
class Golem(Pokemon):
    def __init__(self):
        super().__init__("Golem")
        
class Dewgong(Pokemon):
    def __init__(self):
        super().__init__("Dewgong")
        
class Hypno(Pokemon):
    def __init__(self):
        super().__init__("Hypno")
        
class Cleffa(Pokemon):
    def __init__(self):
        super().__init__("Cleffa")
        
class Cutiefly(Pokemon):
    def __init__(self):
        super().__init__("Cutiefly")

```
