For a user to play Pokemon, the following events occur:

The Pokemon battle starts.

The user attacks its target.

The target loses HP (hit points) and attacks back.

The first one to 0 HP faints and loses.

Hit points determine how much damage a Pokemon can take. For example, Gardevoir can take 220 hit points before it is knocked out at 230. 

Every Pokemon has at least one type of attack. This attack does 30 damage times the amount of energy on both active Pokemon.

Pokemon types dictate whether one Pokemon is strong or weak (i.e., super effective or not very effective) against another, dealing additional or reduced damage as a result. For example, a Fire type is strong against Bug, Steel, Grass, and Ice types, and weak against Rock Fire, Water, and Dragon types. Thus, if you choose Charmander as your Pokemon, which is a Fire type, then you are strong against Bug, Steel, Grass, and Ice types, and weak against Rock Fire, Water, and Dragon types.

If your attack is a weakness of your opponent, then your attack is "super effective", and inflicts 2 times the typical damage done. If your attack is one of your opponent's strengths, then your attack is "not very effective", and delivers half the typical amount of damage done. If your attack is not a weakness or strength of the opponent, then you do a standard amount of damage.



Create a Pokemon class that contains the attributes name, $type_$, HP, damage, strengths, weaknesses, and attacks.

```python
from random import randint

STRENGTHS = {
    "Normal" : ["N/A"],
    "Fire" : ["Grass", "Ice", "Bug", "Steel"],
    "Water" : ["Fire", "Ground", "Rock"],
    "Electric" : ["Water", "Flying"],
    "Grass" : ["Water", "Ground", "Rock"],
    "Ice" : ["Grass", "Ground", "Flying", "Dragon"],
    "Fighting" : ["Normal", "Ice", "Rock", "Dark", "Steel"],
    "Poison" : ["Grass", "Fairy"],
    "Ground" : ["Fire", "Electric", "Poison", "Rock", "Steel"],
    "Flying" : ["Grass", "Fighting", "Bug"],
    "Psychic" : ["Fighting", "Poison"],
    "Bug" : ["Grass", "Psychic", "Dark"],
    "Rock" : ["Fire", "Ice", "Flying", "Bug"],
    "Ghost" : ["Psychic", "Ghost"],
    "Dragon" : ["Dragon"],
    "Dark" : ["Psychic", "Ghost"],
    "Steel" : ["Ice", "Rock", "Fairy"],
    "Fairy" : ["Fighting", "Dragon", "Dark"]
}

WEAKNESSES = {
    "Normal" : ["Rock", "Steel"],
    "Fire" : ["Fire", "Water", "Rock", "Dragon"],
    "Water" : ["Water", "Grass", "Dragon"],
    "Electric" : ["Electric", "Grass", "Dragon"],
    "Grass" : ["Fire", "Grass", "Poison", "Flying", "Bug", "Dragon", "Steel"],
    "Ice" : ["Fire", "Water", "Ice", "Steel"],
    "Fighting" : ["Poison", "Flying", "Psychic", "Bug", "Fairy"],
    "Poison" : ["Poison", "Ground", "Rock", "Ghost"],
    "Ground" : ["Grass", "Bug"],
    "Flying" : ["Electric", "Rock", "Steel"],
    "Psychic" : ["Psychic", "Steel"],
    "Bug" : ["Fire", "Fighting", "Poison", "Flying", "Ghost", "Steel", "Fairy"],
    "Rock" : ["Fighting", "Ground", "Steel"],
    "Ghost" : ["Dark"],
    "Dragon" : ["Steel"],
    "Dark" : ["Fighting", "Dark", "Fairy"],
    "Steel" : ["Fire", "Water", "Electric", "Steel"],
    "Fairy" : ["Fire", "Poison", "Steel"]
}

### ADD MORE ATTACKS
ATTACKS = {
    "Poop": ['Grass'],
    "Bite": ['Fire', 'Water'],
    "Claw": ['Dragon', 'Steel']
}

class Pokemon(object):

    def __init__(self, name, type_, attacks):
        self.name = name
        self.type_ = type_
        self.hp = 50 
        self.strength = type_
        self.attacks = attacks

    def attack(self, target, attack):
        base_damage = self.roll_damage()
        damage = self.modify_damage(base_damage, target, attack)
        print(f"{self.name} attacks {target.name} for {damage} damage!")
        target.take_damage(self, damage)

    def roll_damage(self):
        return randint(1, 10) + randint(1, 10) 

    ### REALLY UNDERSTAND THIS FUNCTION
    ### MAKE A SIMILAR FUNCTION FOR ATTACK AS A STRENGTH
    def attack_is_a_weakness(self, attack, target_weakness):
        stuff_in_common = []
        for attack in ATTACKS[attack]:
            for weakness in target_weakness:
                if attack == weakness:
                    stuff_in_common.append(attack)
        if len(stuff_in_common) > 0:
            return True
        else:
            return False
    
    #### WORK MORE ON CRITICAL DAMAGE
    def modify_damage(self, base_damage, target, attack):
        target_weakness = WEAKNESSES[target.type_]
        target_strength = STRENGTHS[target.type_]
        strength = STRENGTHS[self.type_]
        weakness = WEAKNESSES[self.type_]
        crit_chance = randint(1, 7)
        
        print(f'{self.name} chose to attack {target.name} with {attack}.')
        if self.attack_is_a_weakness(attack, target_weakness):
            damage = base_damage * 2
            print(f"Your {attack} attack was Super Effective!".title())
        else:
            damage = base_damage
            print(f"Your {attack} attack Was Not Very Effective".title())
        if crit_chance > 6:
            damage = base_damage * 2
            print("You landed a critical hit!".title())
        if crit_chance < 3:
            damage = base_damage * 0
            print("Your hit missed the target!".title())
        return damage

    ### HOW TO QUIT GAME IF FAINTS
    def take_damage(self, target, damage):
        self.hp = max(self.hp - damage, 0)
        print(f"{self.name} has {self.hp} health points left.")
        print(f"{target.name} has {target.hp} health points left.")
        if self.hp ==0 :
            raise Exception(f'{self.name} has fainted!')

class Squirtle(Pokemon): 
    def __init__(self, name):
        attacks = ["Poop", "Claw"]
        super().__init__(name, "Water", attacks)
        self.hp = 10
        print(f'Squirtle has {attacks} as attacks.')
        
### ADD MORE SUBCLASSES
class Charmander(Pokemon):
    def __init__(self, name):
        attacks = ["Bite", "Claw"]
        super().__init__(name, "Fire", attacks )
        self.hp = 100
        print(f'Charmander has {attacks} as attacks.')

class Bulbusar(Pokemon):
    def __init__(self, name):
        attacks = ["Bite", "Poop"]
        super().__init__(name, "Grass", attacks )
        self.hp = 50
        print(f'Bulbusar has {attacks} as attacks.')

class Dragalge(Pokemon):
    def __init__(self, name):
        attacks = ["Bite", "Poop"]
        super().__init__(name, "Dragon", attacks )
        self.hp = 200
        print(f'Dragalge has {attacks} as attacks.')
        
bulby = Bulbusar("bulby")
squirtle = Squirtle("squirly")
draggy = Dragalge("draggy")
print(squirtle.hp)
```
