```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

1.We are going to explore the top 100 movie releases of 2017 using the dataset here:

http://www.boxofficemojo.com/yearly/chart/?yr=2017&p=.htm

First, read in the 100 movie table using pd.read_html.

```python
table = pd.read_html('http://www.boxofficemojo.com/yearly/chart/?yr=2017&p=.htm')[2]
```

2.Only rows 6 through 106 and the odd columns 1,3,5 and 7 contain valuable info that we will use. Save this reduced table to a dataframe called movies. Then reset the index. Be sure the changes save!

```python
movies = table.iloc[6:106,[1,3,5,7]]
movies = movies.reset_index(drop=True)
movies.head()
```

3.Rename the columns to be movie, total gross, opening gross, open date.

```python
movies = movies.rename(index=str, columns={1: 'movie',
                                           3: 'total gross',
                                           5: 'open gross',
                                           7: 'open date'})
movies.head()
```

4.You can strip the dollar signs and commas from the total gross using list comprehensions. Study the code carefully below and run it:

```python
movies['total gross'] = [x.strip('$') for x in movies['total gross']]
movies['total gross'] = [x.replace(',', '') for x in movies['total gross']]
```

5.Strip the dollar signs from the open gross column.

```python
movies['open gross'] = [x.strip('$') for x in movies['open gross']]
movies['open gross'] = [x.replace(',','') for x in movies['open gross']]
```

6.Plot total gross on the x axis and open gross on the y-axis.

```python
plt.plot(movies['total gross'],movies['open gross'],'.')
plt.xlabel('gross')
plt.ylabel('open')
```

7.Change the gross columns to integer data types.

```python
movies['open gross'] = movies['open gross'].astype(int)
movies['total gross'] = movies['total gross'].astype(int)
```

8.Create another column called ratio that is given by the opening gross divided by the total gross. Do it in one line.

```python
movies['ratio'] =  movies['open gross']/movies['total gross']
```

9.Sort the ratio from largest to smallest to find the biggest "sleeper" hit (meaning, the movie that didn't make that much in its opening weekend but made a lot of money later.)

```python
movies.sort_values(by = 'ratio')
```

10.Plot a histogram of the ratios. Make the rwidth by 0.9 and make the bins start at 0 and end at 1 in increments of 0.1.

```python
plt.hist(movies['ratio'],rwidth=0.9, bins=np.arange(0,1,.1))
```

13.Loop through all of the movies in your pandas dataframe. Each time, use Selenium to search for the movie's director. Update your pandas dataframe to include that director within the loop. 

Two things to keep in mind before you start:

1.You'll want to use a try/except because a few searches will mess up. Print the "bad" movies and their index to the screen as you loop through so you know which ones to search for manually when your program finishes.

2.You'll want to add in some 1 second sleeps in your code in order to make it run properly.

3.The website will throttle you if try to search for too many at once. Try doing it in chunks of 25 movies at a time and then take a few minutes break.

```python
import time

for i in range(54,len(movies)):
    if movies['director'][i]=='':
        try:
            chromedriver = "/usr/local/bin/chromedriver"
            driver = webdriver.Chrome(chromedriver)
            url = "http://www.boxofficemojo.com/"
            driver.get(url)
            search = driver.find_element_by_xpath('//*[@id="leftnav"]/li[2]/form/input[1]')
            search.send_keys(movies['movie'][i])
            box = driver.find_element_by_xpath('//*[@id="leftnav"]/li[2]/form/input[2]')
            box.click()
            time.sleep(1)
            movie = driver.find_element_by_xpath('//*[@id="body"]/table[2]/tbody/tr/td/table[2]/tbody/tr[2]/td[1]/b/font/a')
            movie.click()
            time.sleep(2)
            director = driver.find_element_by_xpath('//*[@id="body"]/table[2]/tbody/tr/td/table[2]/tbody/tr[2]/td/table/tbody/tr/td[1]/table/tbody/tr/td[2]/div/div[2]/table/tbody/tr[1]/td[2]/font/a').text
            print(movies['movie'][i])
            movies['director'][i] = director
            time.sleep(1)
            driver.close()
        except:
            print('BAD', i,movies['movie'][i])
        
```
