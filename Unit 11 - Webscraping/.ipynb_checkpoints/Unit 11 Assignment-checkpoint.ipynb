Retrieving web pages using the requests package
---
<a class="anchor" id="requests"></a>

The requests library allows you to send HTTP requests using Python. Using requests, you can treat a web page much like a file. You simply indicate which web page you would like to retrieve and requests handles all of the HTTP protocol and header details.
The code below reads the romeo.txt file from the website http://data.pr4e.org/romeo.txt :

```python
import requests

url = 'http://data.pr4e.org/romeo.txt'
response = requests.get(url)
print(response.text)
```

So response.text prints out the text from the web page. What does response print out? The response from the website. A response of 200 means that you were able to communicate with the website successfully:

```python
print(response)
```

There are tons of different codes you can receive. You've probably received a 403 Forbidden error or a 404 Not Found error when you've gone to a website. All the codes are here:

https://en.wikipedia.org/wiki/List_of_HTTP_status_codes

What if you wanted to print all of the header details that come along with sending an HTTP request? Type:

```python
print(response.headers)
```

The Romeo text is short but many web pages will contain a lot of info. What if instead of printing all of the text at once you want to print it line by line? We'll need to break up each line by the newline character: 

```python
import requests

url = 'http://data.pr4e.org/romeo.txt'
sentences = requests.get(url).text.split('\n')

for sentence in sentences:
    print(sentence)
```

Is there a more Pythonic way of printing each line? Yes. Use the following:

```python
import requests

url = 'http://data.pr4e.org/romeo.txt'
r = requests.get(url)

for line in r.iter_lines():
    print(line)
```

What do those b's stand for? Bytes. To convert from a byte array to strings that you are used to, use decode:

```python
import requests

url = 'http://data.pr4e.org/romeo.txt'
r = requests.get(url)

for line in r.iter_lines():
    print(line.decode())
```

What if we want to print a list of the most frequently used words in the text? We've done this before in previous units using a dictionary and then a sorted list of tuples:

```python
import requests

url = 'http://data.pr4e.org/romeo.txt'
r = requests.get(url)

counts = {}

for line in r.iter_lines():
    words = line.decode().split() 
    for word in words:
        counts[word] = counts.get(word,0) + 1 

counts_list=[]
for key, val in counts.items():
    counts_list.append((val,key))

counts_list.sort(reverse = True)
print(counts_list)
```

The faculty names seem to be located below a line that contains the term "FullName." We can make a loop to print out just the names by creating a boolean variable called "fullname" that keeps track of whether FullName is in the line or not. If it is, we'll print out the next line:

```python
url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

fullname = False
for line in r.iter_lines():
    line = line.decode()
    if fullname == True:
        print(line)
        fullname = False
    if 'FullName' in line:
        fullname = True
```

Web Scraping using BeautifulSoup
---
<a class="anchor" id="scraping"></a>

One of the common uses of the requests capability in Python is to scrape the web. Web scraping is when we write a program that pretends to be a web browser and retrieves pages, then examines the data in those pages looking for patterns.

As an example, a search engine such as Google will look at the source of one web page and extract the links to other pages and retrieve those pages, extracting links, and so on. Using this technique, Google spiders its way through nearly all of the pages on the web.


Google also uses the frequency of links from pages it finds to a particular page as one measure of how “important” a page is and how high the page should appear in its search results. 

BeautifulSoup is one Python package that helps us to scrape the web. BeautifulSoup tolerates highly flawed HTML and still lets you easily extract the data you need.

If you have Anaconda installed on your computer, you probably already have it, so skip to trying to run the cell below.

If you don't have BeautifulSoup, to download it, type "conda install beautifulsoup4" or "pip install beautifulsoup4" into your terminal. (You can find your terminal under Applications - Utilities on a Mac). If that doesn't work, download it directly from this website: https://www.crummy.com/software/BeautifulSoup/

Restart your kernel and then try running the following cell to see if you have BeautifulSoup installed:

```python
from bs4 import BeautifulSoup
```

Okay, let's first view what information is contained on the following page:

```python
import requests
url = 'http://www.dr-chuck.com/page1.htm'
r = requests.get(url)
print(r.text)
```

Running this webpage through Beautiful Soup gives us a BeautifulSoup object, which represents the document as a nested data structure. If we print the soup object, it will look like your requests output above:

```python
url = 'http://www.dr-chuck.com/page1.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

print(soup)
```

However, this soup data structure will contain many other pieces of nested info. For example, if you only wanted all of the the paragraph tags, you could type:

```python
soup.find_all('p')
```

Or as a shortcut, you could type:

```python
print(soup('p'))
```

Notice that even though there is only one paragraph tag above, these commands return it in a list with brackets around in, in case there are multiple paragraph tags. If you wanted to find just the **first** web anchor tag, you could type:

```python
soup.find('a')
```

Or more simply:

```python
soup.a
```

The anchor tag itself actually contains a lot of nested info. We'll often want to grab either just the URL or just the text contained between the web anchor tab:

```python
tag = soup.find('a')

print(f"TAG: {tag} \n")
print(f"URL: {tag.get('href')} \n")
print(f"TEXT: {tag.get_text()} \n")
```

Notice that using HTML, headers are contained between the "h1" terms, web link tags are contained within the "a" terms, and paragraphs within the "p" terms. If we want to add indentation to see the structure of the webpage better, we can use "prettify":

```python
print(soup.prettify())
```

Notice that in the tag above, we obtained the text "Second page" from using .get_text(), but we also could have used .contents:

```python
print(tag.contents[0])
```

As another example, consider this OES webpage:

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

paragraphs = soup('p')
for paragraph in paragraphs:
    print(paragraph)
```

Notice that the OES mission is contained in the third set of paragraph tags, so we'll narrow our print statement down to:

```python
paragraphs = soup('p')
print(paragraphs[2])
```

To print out just the content between the paragraph tags, we'll type:

```python
print(paragraphs[2].contents[0])
```

As another example, suppose that you read in a webpage containing the contents below:

```python
html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<a href="http://example.com/alice" class="maincharacter" id="link">Alice</a>,

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p href="http://example.com/other" class="sister" id="link4">Were there any other sisters?</p>;

<p class="story">...</p>
"""

from bs4 import BeautifulSoup
soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.prettify())
```

Notice that there are four anchor tags:

```python
soup.find_all('a')
```

To limit our scope to just the tags with the class "sister", we can type the following (note that class has an underscore attached since class is a reserved word in Python:

```python
soup.find_all(class_ = "sister")
```

If we wanted to limit our scope to just the three sisters' names, we could limit our search to only the a tags and ignore the p tag:

```python
soup.find_all("a", class_ = "sister")
```

If we wanted to print each sister's name, we could type:

```python
for tag in soup.find_all("a", class_ = "sister"):
    print(tag.get_text())
```

If we wanted to print their urls, we could type:

```python
for tag in soup.find_all("a", class_ = "sister"):
    print(tag.get('href'))
```

### Finding Names

Let's return to our OES faculty example. Look carefully at the teachers' names. Where are they stored? Inside "h3" tags. Do a Control - Find to see Carmen Boyle's name below inside h3 tags:

```python
import requests

url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

for i,line in enumerate(r.iter_lines()):
    if line:
        if 620 < i < 700: #i'm only printing out part of the document since
                            #it is so large for easier viewing
            line = line.decode()
            print(i,line)
```

Let's try printing out just the h3 info using BeautifulSoup:

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

headers = soup('h3')
print(headers)
```

That's still not quite as pretty as we'd like, so we notice that the actual names are stored inside the anchor tag's text, and then we can strip away any extra whitespace:

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

headers = soup('h3')

for header in headers:
    print(header.get_text().strip())
```

Notice that by using BeautifulSoup, we didn't need to create a loop to search for "Full Name" as we did at the beginning of this unit.

### Finding departments

Suppose we wanted to get what department each faculty member works in. First we notice that their departments are inside 'div class="fsDepartments"':

```python
import requests

url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

for i,line in enumerate(r.iter_lines()):
    if line:
        if 620 < i < 700: #once again, this line is only for ease of viewing
            line = line.decode()
            print(i,line)
```

Let's first try printing out all of the tags labeled "div":

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

divs = soup('div')
print(divs)
```

That's still too much info. To narrow down our search, we can use BeautifulSoup to find the classes labeled "fsDepartments":

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

depts = soup.findAll(class_ = "fsDepartments")

for dept in depts:
    print(dept)
```

That's still a bit too messy. After playing around, we see that what we really want is the text contained in contents[2]:

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.oes.edu/academics/upper-school/faculty-staff'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

depts = soup.findAll(class_ = "fsDepartments")

for dept in depts:
    print(dept.contents[2].strip())
```

Using the Developer console to help you find stuff
---
<a class="anchor" id="javascript"></a>

Let's use Beautiful Soup and requests to find the temperature for Portland using www.weatherunderground.com. First let's read in the web page using requests.

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.wunderground.com/weather-forecast/97217'
response = requests.get(url)

print(response.text)

```

Okay, this is really ugly. We'll have a hard time finding the information we need from here. Here's a trick to find it more easily:

First, go the website http://www.wunderground.com/weather-forecast/97223 

Then, right click on the current temperature.

You should see something like this:
<img src="weatherimage.jpg" style="width: 400px;"/>

Then, choose Inspect. This will cause the developer console to pop up:
<img src="weatherimage2.jpg" style="width: 700px;"/>

**Sometimes you will need to right click on the desired image again to get the specific blue line that we are interested in.**

What is it telling us? Well, the blue highlighted section is telling us the part of the html code that contains the temperature info we clicked on. If you look closely, it is described by a 'wu-value wu-value-to' class tag. In order to extract the temperature, we can use BeautifulSoup. 


```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.wunderground.com/weather-forecast/97217'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

temp = soup.find(class_ = 'wu-value wu-value-to').get_text()
print('Temp:', temp)
```

What if we wanted to obtain the description of the weather? We could click on the sun/cloud/snow icon:

<img src="weatherimage3.jpg" style="width: 200px;"/>

If we click on Inspect, we see that it is located under the "condition-icon small-6 medium-12 columns" class tag. 

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.wunderground.com/weather-forecast/97217'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

conditions = soup.find(class_ = 'condition-icon small-6 medium-12 columns').get_text()
print('Conditions:', conditions)
```

Many times, we will want to scrape the web and put the data that we get into a Pandas dataframe. For example, view the seven day forecast on the website here:

https://forecast.weather.gov/MapClick.php?lat=45.44763999999998&lon=-122.76902000000001#.WqL4D5MbOu4

Right click on the seven day forecast to see that it is located under the tag "col-sm-10 forecast-text":

<img src="weathersevenday.jpg" style="width: 400px;"/>

Notice that in this case, there are multiple weather descriptions stored under the "col-sm-10 forecast-text" tag. In this case, we'll need to use findAll instead of find, since we are looking for more than one thing. This will return all of the results in a list. We can use .text to get the text from each result:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://forecast.weather.gov/MapClick.php?lat=45.44763999999998&lon=-122.76902000000001#.WqL4D5MbOu4'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

descriptions = soup.findAll(class_='col-sm-10 forecast-text')
for description in descriptions:
    print(description.get_text())
```

If we want to store these descriptions in a Pandas dataframe, we'll first want to store these descriptions in a list:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://forecast.weather.gov/MapClick.php?lat=45.44763999999998&lon=-122.76902000000001#.WqL4D5MbOu4'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

descriptions = soup.findAll(class_='col-sm-10 forecast-text')
forecasts = []
for description in descriptions:
    forecasts.append(description.get_text())
```

Let's do the same thing to grab the days of the week:

```python
import requests
from bs4 import BeautifulSoup

url = 'https://forecast.weather.gov/MapClick.php?lat=45.44763999999998&lon=-122.76902000000001#.WqL4D5MbOu4'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

descriptions = soup.findAll(class_='col-sm-2 forecast-label')
days = []
for description in descriptions:
    days.append(description.get_text())
```

Finally, we can convert these two lists into a pandas dataframe:

```python
import pandas as pd
weather = pd.DataFrame({
        "days": days, 
        "forecasts": forecasts, 
    })
weather
```

Using Pandas to Webscrape
---
Pandas has a super cool built in functionality to read tables from websites! It doesn't work on all tables, but when it does, it saves you a heck of a lot of time!
<a class="anchor" id="http"></a>

First, go to this website listing NFL Super Bowl champions:

https://en.wikipedia.org/wiki/List_of_Super_Bowl_champions

There are several tables on the website. We can print all of them in just a few lines of code!

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_Super_Bowl_champions'
tables = pd.read_html(url)
print(tables)
```

There are several tables contained within this printout. It looks like the table I'm really interested in is the second one. Let's print that one:

```python
df = tables[1]
df
```

Let's rename our columns the first row of the dataframe and then drop the first row:

```python
df.columns = df.iloc[0]
df.drop(0, inplace = True)
df
```

Obviously, the data would take a bit of cleanup, but it's a very quick way to get what you need!
