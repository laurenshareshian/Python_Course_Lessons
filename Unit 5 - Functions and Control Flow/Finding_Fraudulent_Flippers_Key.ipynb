### Clue if No One gets the answer
You can quantify intuitive notions of randomness by defining a run as a sequence of consecutive trials that result in the same value. HHHHTTTT has two runs: a run of heads followed by a run of tails. HTTTHHTT has 4 runs. HTHTHTHTHTHTHTH has 15 runs: eight runs of heads and seven runs of tails.

It turns out that you can calculate the expected number of runs in a random sequence that has n heads and m tails. The expected number of runs is E(R) = 2nm / (n+m) + 1. If a sequence has n=16 heads and m=14 tails, the expected number of runs is 15.9. So, the further a string of flips is from the expected number of runs, the less likely that list of flips is actually random.

```python
def expected_runs(flips):
    '''
    This take in a string or list of flips
    returns a tuple of (runs in list, expected runs, abs(difference of the two))
    '''
    runs = 0
    last = ''
    for flip in flips:
        if flip != last:
            runs +=1
            last = flip
    h = flips.count('H')
    t = flips.count('T')
    expected = ((2*h*t)/(h+t))+1
    return (runs, expected, abs(runs-expected))
```
